```python
def airMolecWt():
    return un.Mass.g / un.Amount.Mol * 28.9647

def airDensity(pressure: um.Quantity, temperature: um.Quantity):
    # Curvefit of PPDS fluid data
    rhoStd = 1.2922 * un.Density.kg_m3
    pStd = 1.00 # * un.Pressure.atm
    tStd = 273.15 # * un.Temperature.degK

    if (pressure.Similar(un.Pressure.atm)):
        p_atm = pressure.Value(un.Pressure.atm)
 
    if (temperature.Similar(un.Temperature.degK)):
        t_k = temperature.Value(un.Temperature.degK)
    
    f = p_atm / pStd * tStd / t_k
    
    return rhoStd * f
```



Calculation of internal bellows pressure may require:<br>

- Dalton's Law of Partial Pressure<br>
- Raoult's Law (for mixed fills)<br>
- Ideal Gas Law (for trapped air)<br>

Assume the bellows is filled at 1.0 atm and 70 °F<br>

Required Parameters:<br>
- Internal volume at the filling condition<br>
- Internal volume at the on-seat condition<br>
- Bellows spring rate<br>
- Bellows mean effective area = pi * (od + id)^2 / 16<br>
- Seat diameter<br>
- Pressure downstream of the seat<br>
- Bellow fill components, chemical properties & amounts<br>

```python
utf = un.Temperature.degF
utdf = un.Temperature.deltaF
uvp = un.Pressure.atm
upsig = un.Pressure.psig
upsia = un.Pressure.psia
ugd = un.Density.g_cm3
uld = un.Density.g_cm3
uvol = un.Volume.ml

# parameters for p/n 2751500
vol_at_fill = 0.2471 * un.Volume.inCu # bellows internal volume during the filling process, 
vol_on_seat = 0.4159 * un.Volume.inCu # bellows internal volume at the on-seat condition
stroke = 0.170 * un.Length.inch
spr_rate = un.Force.lbf / un.Length.inch * 36.0
mea = 0.994 * un.Length.inch.squared()
seat_dia = 0.3125 * un.Length.inch
seat_area = seat_dia.squared() * np.pi / 4.0

dp_to_stroke = stroke * spr_rate / mea
print (f'dp to stroke: {dp_to_stroke.ValueAsStr(un.Pressure.psi)}')

t_amb = 70.0 * utf
p_amb = 1.0 * un.Pressure.atm
R_air = 0.287 * un.SpHeatCap.kJ_kgK
R_univ = un.Energy.J / un.Amount.Mol / un.Temperature.degK * 8.3145

water = h2o.WaterIAPWS97()

p_vap = VaporPressure(t_amb)
rho_fill_vap = p_vap * MolecWt() / R_univ / t_amb
rho_fill_liq = SatLiquidDensity(t_amb)

m_fill1 = 0.11 * un.Mass.g
m_fill2 = 0.0 * un.Mass.g
m_fill = m_fill1 + m_fill2

vol_liq_at_fill = (m_fill - vol_at_fill * rho_fill_vap) / (rho_fill_liq - rho_fill_vap)

#### completely filled bellows
# vol_liq_at_fill = vol_at_fill
# m_fill = vol_liq_at_fill * rho_fill_liq 

vol_vap_at_fill = vol_at_fill - vol_liq_at_fill

#### remove air from bellows
# p_amb = p_vap

p_air = p_amb - p_vap
m_air = p_air * vol_vap_at_fill / R_air / t_amb
x_at_fill = vol_liq_at_fill / vol_at_fill
print (f'\nmass of air: {m_air.ValueAsStr(un.Mass.mg)}')
print (f'mass of fill: {m_fill.ValueAsStr(un.Mass.mg)}')
print (f'density of fill vap: {rho_fill_vap.ValueAsStr(un.Density.g_cm3)}')
print (f'density of fill liq: {rho_fill_liq.ValueAsStr(un.Density.g_cm3)}')
print (f'volume of fill vap: {vol_vap_at_fill.ValueAsStr(un.Volume.cc)}')
print (f'volume of fill liq: {vol_liq_at_fill.ValueAsStr(un.Volume.cc)}')
print (f'fraction of liquid volume: {x_at_fill.ValueAsStr(un.Dimensionless.none)}\n')

xc = []
yc = []
xo = []
yo = []

xcond = []
ycond = []

xfill = []
yfill = []

print(f'{"t fill":>10} {"press air":>12} {"press fill":>12} {"press tot":>12} {"sub close":>10} {"sat press c":>12} {"sat temp c":>10} {"sub open":>10}  {"sat press o":>12} {"sat temp o":>10} {"x fill":>10} {"vol liq":>10} {"vol vap":>10}')
print(f'{utf.Symbol:>10} {upsia.Symbol:>12} {upsia.Symbol:>12} {upsia.Symbol:>12} {utdf.Symbol:>10} {upsig.Symbol:>12} {utf.Symbol:>10} {utdf.Symbol:>10} {upsig.Symbol:>12} {utf.Symbol:>10} {"":>10} {uvol.Symbol:>10} {uvol.Symbol:>10}')
print(f'{"_____":>10} {"_____":>12} {"_____":>12} {"_____":>12} {"_____":>10} {"_____":>12} {"_____":>10} {"_____":>10} {"_____":>12} {"_____":>10} {"_____":>10} {"_____":>10} {"_____":>10}')

t_fill = 212.0 * utf
tIncr = 1.0 * utdf

for i in range(500):
    p_vap = VaporPressure(t_fill)
    rho_vap_fill = p_vap * MolecWt() / R_univ / t_fill
    rho_liq_fill = SatLiquidDensity(t_fill)

    vol_liq = (m_fill - vol_on_seat * rho_vap_fill) / (rho_liq_fill - rho_vap_fill)
    vol_vap = vol_on_seat - vol_liq
    m_vap_fill = vol_vap * rho_vap_fill
    x_fill = m_vap_fill / m_fill 
    if (x_fill.Value(un.Dimensionless.none) > 1.0):
        print(f'ran out of fill')
   
    p_air = m_air * R_air * t_fill / vol_vap
    p_fill = p_air + p_vap

   # extend the bellows to close the valve (go on seat)
    p_cond_close = p_fill - dp_to_stroke
    fp = water.SetSaturation(h2o.SatType.SatPress).SetQuality(100.0).SetCond(press=p_cond_close).Eval()
    t_sat_close = water.Properties.Temp
    sub_close = t_sat_close - t_fill

    # Pull the bellows off the seat with p_cond_close acting across seat area
    p_cond_open = (stroke * spr_rate - p_fill * mea) / (seat_area - mea)
    
    fp = water.SetSaturation(h2o.SatType.SatPress).SetQuality(100.0).SetCond(press=p_cond_open).Eval()
    t_sat_open = water.Properties.Temp
    sub_open = t_sat_open - t_fill

    print(f'{t_fill.Value(utf):10.1f} {p_air.Value(upsia):12.4f} {p_vap.Value(upsia):12.4f} {p_fill.Value(upsia):12.4f} {sub_close.Value(utdf):10.1f} {p_cond_close.Value(upsig):12.4f} {t_sat_close.Value(utf):10.1f} {sub_open.Value(utdf):10.1f} {p_cond_open.Value(upsig):12.4f} {t_sat_open.Value(utf):10.1f} {x_fill.Value(un.Dimensionless.none):10.4f} {vol_liq.Value(uvol):10.7f} {vol_vap.Value(uvol):10.7f}')

    xc.append(p_cond_close.Value(upsig))
    yc.append(sub_close.Value(utdf))

    xo.append(p_cond_open.Value(upsig))
    yo.append(sub_open.Value(utdf))

    xcond.append(p_cond_open.Value(upsig))
    ycond.append(t_sat_open.Value(utf))

    xfill.append(p_vap.Value(upsig))
    yfill.append(t_fill.Value(utf))

    t_fill += tIncr
    if (p_cond_close.Value(upsig) > 125.0):
        break

fig, ax = plt.subplots(1, 1, figsize=(16,9))
ax.plot(xc, yc, label='subcool @ close')
ax.plot(xo, yo, label='subcool @ open')
# ax.plot(xcond, ycond)
# ax.plot(xfill, yfill)

ax.set(xlabel='steam pressure (psig)', ylabel='subcooling (°F)',
       title='Bellows 2751500 subcooling')
ax.grid()
leg = plt.legend(loc='right')
plt.show()
```
